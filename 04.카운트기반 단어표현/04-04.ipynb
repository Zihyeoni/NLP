# **04-04 TF-IDF (Term Frequency-Inverse Document Frequency)**

---



DTM 내에 있는 각 단어에 대한 중요도 계산. 기존의 DTM 보다 많은 정보를 고려하여 문서들 비교 가능

## **1. TF-IDF (단어 빈도-역 문서 빈도)**

---



단어의 빈도와 역 문서 빈도(문서의 비도에 특정식을 취함)를 사용해 DTM 내의 각 단어들마다 중요한 정도를 가중치로 주는 방법. 우선 DTM을 만든 후, TF-IDF 가중치 부여

TF-IDF = TF * IDF (문서: d, 단어: t, 문서 총 개수: n)

### **(1) tf(d,t) : 특정 문서 d에서의 특정 단어 t의 등장 횟수**

TF: 사실상 DTM의 예제에서 각 단어들이 가진 값들

### **(2) df(t) : 특정 단어 t가 등장한 문서의 수**

오직 특정 단어 t가 등장한 문서의 수에만 관심 가짐

DTM에서 바나나가 문서2와 문서3에서 등장 -> 바나나의 df=2

바나나란 단어가 문서2에서 100번, 문서3에서 200번 등장했더라도 바나나의 df는 2
