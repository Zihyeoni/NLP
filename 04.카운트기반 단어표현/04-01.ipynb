## **1. 단어의 표현 방법**

---





*   국소 표현(Local Representation) =: 이산 표현(Discrete)
    
    해당 단어 그 자체만 보고, 특정값을 맵핑하여 단어 표현
*   분산 표현(Distributed Representation) =: 연속 표현(Continuous)

    그 단어를 표현하고자 주변을 참고하여 단어 표현


ex) puppy(강아지), cute(귀여운), lovely(사랑스러운)

국소 표현: 각 단어에 1번, 2번, 3번 등과 같은 숫자 mapping (단어의 의미, 뉘앙스 표현 X)

분산 표현: puppy 근처에는 cute, lovely가 자주 등장하므로, puppy는 cute, lovely한 느낌이다 로 단어 정의 (단어의 뉘앙스 표현 O)
