# **03-05 펄플렉서티 (Perplexity, PPL)**

---



두 개의 모델 A, B의 성능을 비교한다면 오타 교정, 기계 번역 등의 평가에 투입해볼 수 있음. 그러나 일일히 모델들에 대해서 실제 작업을 시켜보고 정확도를 비교하는 작업은 공수가 많이 드는 작업임. 만약 비교할 모델이 두 개 그 이상의 수라면 시간은 그만큼 배로 늘어날 수 있음

펄플렉서티(perplexity) : 조금은 부정확할 수 있어도 테스트 데이터에 대해 빠르게 식으로 계산함. 모델 내에서 자신의 성능을 수치화하여 결과를 내놓음

## **1. 언어 모델의 평가 방법(Evaluation metric) : PPL**

---

## **2. 분기 계수 (Branching factor)**

---



선택할 수 있는 가능한 경우의 수. PPL은 분기계수임. 이 언어 모델이 특정 시점에서 평균적으로 몇 개의 선택지를 가지고 고민하고 있는지를 의미함. 같은 테스트 데이터에 대해서 두 언어 모델의 PPL을 각각 계산 후에 값을 비교하면, 두 언어 모델 중 PPL이 더 낮은 모델의 성능이 더 좋다고 볼 수 있음

> image.png

단, 주의할 점은 PPL 값이 낮다는 것은 테스트 데이터 상에서 높은 정확도를 보인다는 것이지, 사람이 직접 느끼기에 좋은 언어 모델이라는 것을 반드시 의미하진 않음. 또한 두 개 이상의 언어 모델을 비교할 때는 정량적으로 양이 많고, 도메인에 알맞은 동일한 테스트 데이터를 사용해야 신뢰도가 높음
