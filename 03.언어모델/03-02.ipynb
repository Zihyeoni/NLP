## **2. 문장에 대한 확률**

---



각 단어는 문맥이라는 관계로 인해 이전 단어의 영향을 받아 나온 단어이고, 모든 단어로부터 하나의 문장이 완성됨.

문장의 확률을 구하고자 조건부 확률을 사용함. 조건부 확률의 일반화 식을 문장의 확률 관점에서 다시 적어보면, 문장의 확률은 각 단어들이 이전 단어가 주어졌을 때 다음 단어로 등장할 확률의 곱으로 구성됨

> image.png

문장 'An adorable little boy is spreading smiles'의 확률, ***P(An adorable little boy is spreading smiles)***

## **3. 카운트 기반의 접근**

---



SLM은 카운트에 기반하여 이전 단어로부터 다음 단어에 대한 확률을 계산함

An adorable little boy가 나왔을 때 is가 나올 확률, ***P(is|An adorable little boy)***

> image.png

예를 들어 기계가 학습한 코퍼스 데이터에서 An adorable little boy가 100번 등장했는데 그 다음에 is가 등장한 경우가 30번이라고 한다면, ***P(is|An adorable little boy)***는 30%

## **4. 카운트 기반 접근의 한계 - 희소 문제(Sparsity Problem)**

---



언어 모델은 실생활에서 사용되는 언어의 확률 분포를 근사 모델링함. 이를 실제 자연어의 확률 분포, 현실에서의 확률 분포라고 명칭함

언어 모델의 목표: 기계에게 많은 코퍼스를 훈련시켜서 언어 모델을 통해 현실에서의 확률 분포를 근사하는 것

그런데 카운트 기반으로 접근하려 한다면 갖고있는 코퍼스(corpus),즉 기계가 훈련하는 데이터는 방대한 양이 필요함

> image.png

예를 들어 P(is|An adorable little boy)를 구하는 경우, 기계가 훈련한 코퍼스에 An adorable little boy is라는 단어 시퀀스가 없었다면 이 단어 시퀀스에 대한 확률은 0이 됨. 또는 An adorable little boy라는 단어 시퀀스가 없었다면 분모가 0이 되어 확률은 정의되지 않음.

이와 같이 충분한 데이터를 관측하지 못해 정확히 모델링하지 못하는 문제를 **희소 문제(sparsity problem)**라고 함

위 문제를 완화하는 방법으로 n-gram 언어 모델이나 스무딩, 백오프와 같은 여러가지 일반화 기법이 존재함. 하지만 근본적인 해결책이 되지 못한 채 언어 모델의 트렌드는 통계적 언어 모델에서 인공 신경망 언어 모델로 넘어가게 됨
