---



image.png

**훈련 데이터**: 머신 러닝 모델 학습

**검증 데이터**: 모델의 성능 조정 (성능 평가 X)
- 모델이 훈련 데이터에 과적합(overfitting) 되고있는지 판단, 하이퍼파라미터의 조정

> image.png



훈련 데이터로 훈련시킨 모델은 검증 데이터를 사용해 정확도 검증, 하이퍼파라미터를 **튜닝(tuning)**함 (높은 정확도를 얻도록 하이퍼파라미터 값 바꿔봄). 튜닝하는 과정에서 모델은 검증 데이터의 정확도 높이는 방향으로 점차적 수정됨

튜닝 과정 모두 끝나면, 테스트 데이터로 모델 최종 평가함. (훈련 데이터: 공부를 위한 문제지, 검증 데이터: 모의고사, 데스트 데이터: 수능 시험)

## **2. 분류(Classification)와 회귀(Regression)**

---



이번 챕터에서 다루는 머신러닝 기법
1. 선형 회귀(Lineare Regression): 대표적 회귀 문제
2. 로지스틱 회귀(Logistic Regression): 대표적 분류 문제

분류
*   이진 분류 (Binary Classification)
*   다중 클래스 분류 (Multi-Class Classification)





### **1) 이진 분류 문제 (Binary Classification)**

주어진 입력에 대해 두 개의 선택지 중 하나의 답을 선택해야 하는 경우

예) 시험 성적표를 보고 최종 합격/불합격 판단하는 문제, 메일을 보고 정상 메일/스팸 메일 판단하는 문제 등

### **2) 다중 클래스 분류 (Multi-class Classification)**

주어진 입력에 대해 세 개 이상의 선택지 중에서 답을 선택해야 하는 경우

예) 과학, 영어, IT, 만화, 학습지 총 5가지 책장이 있음. 이때 새 책이 입고되어 다섯 개의 책장 중 분야에 맞는 책장에 책을 넣어야 할 경우

### **3) 회귀 문제 (Regression)**

정답이 몇 개의 정해진 선택지 중 정해져 있는 경우가 아닌, 어떠한 연속적인 값의 범위 내에서 예측값이 나오는 경우

예) 시계열 데이터(Time Series Data)를 이용한 주가 예측, 생산량 예측, 지수 예측 등

## **3. 지도 학습과 비지도 학습**

---



머신 러닝

*   지도 학습
*   비지도 학습
*   강화 학습



### **1) 지도 학습 (Supervised Learning)**

정답인 레이블(Label)과 함께 학습, (레이블 = 실제값 = y)

자연어 처리는 대부분 지도 학습에 속함

### **2) 비지도 학습 (Unsupervised Learning)**

데이터에 별도의 Label 없이 학습

ex) 텍스트 처리 분야의 토픽 모델링 알고리즘, LSA & LDA

### **3) 자기지도 학습 (Self-Supervised Learning, SSL)**

Label이 없는 데이터가 주어질 때, 모델이 학습을 위해 스스로 데이터로부터 Label을 만들어 학습

ex) Word2Vec(워드 임베딩 알고리즘), BERT(언어 모델의 학습 방법)

### **3) 정확도 (Accuracy)**

전체 예측한 데이터 중에서 정답을 맞춘 것에 대한 비율

> image.png

그러나, Accuracy로 성능 예측하는 것이 적절하지 않은 경우 존재

ex1) 200일 동안 총 6일만 비가 왔다고 가정, 모델은 200일 내내 날씨가 맑았다고 예측.

> 이 모델은 200번 중 6회 틀렸으므로 정확도는 97% (194/200=0.97). 하지만 정작 비가 온 날은 하나도 못 맞춤

ex2) 메일 100개 중 스팸 메일은 5개일 때, 스팸 메일 분류기는 모두 정상 메일이라고 탐지.

> 정확도는 95%이지만, 정작 스팸 메일은 하나도 못 찾음

이렇게 실질적으로 더 중요한 경우의 데이터가 전체 데이터에서 너무 적은 비율을 차지한다면, 정확도는 좋은 측정 지표가 될 수 없음. 이런 경우 **F1-Score** 사용 (개체명 인식 챕터에서 설명)

## **6. 과적합(Overfitting)과 과소 적합(Underfitting)**

---



### **적합 (fitting)**

머신 러닝에서 학습 또는 훈련하는 과정 (모델이 주어진 데이터에 대해 적합해져가는 과정)

케라스에서 기계를 학습시킬 때 fit() 호출

#### **1) 과적합 (Overfitting)**

훈련 데이터를 과하게 학습한 경우

기계가 훈련 데이터에 대해서만 과하게 학습하면 테스트 데이터나 실제 서비스에서는 정확도가 낮은 현상 발생

> image.png

훈련 데이터에 대해서는 오차(loss)가 낮지만, 테스트 데이터에 대해서는 오차가 커짐.

위 그래프에서 보여주듯, epoch가 증가할수록 test data에 대한 오차가 점차 증가. train data에 대한 정확도는 높지만, test data는 정확도가 낮은 상황이라고 말할 수 있음.

이런 상황을 방지하기 위해서는 test data의 오차가 증가하기 전이나, 정확도가 감소하기 전에 훈련을 멈추는 것이 바람직함

#### **2) 과소적합 (Underfitting)**

테스트 데이터의 성능이 올라갈 여지가 있음에도 훈련을 덜 한 상태

훈련 자체가 부족한 상태이므로 epoch(훈련 횟수)가 지나치게 적으면 발생 (훈련 데이터의 정확도가 낮음)

딥 러닝할 때 과적합 막을 수 있는 방법: Dropout, Early Stopping 등
