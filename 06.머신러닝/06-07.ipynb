# **06-07 다중 입력에 대한 실습**

---



독립 변수 x가 2개 이상인 경우

= 입력 벡터의 차원이 2 이상

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
```

## **1. 다중 선형 회귀**

---



image.png

train: 상위 5개의 데이터

test: 나머지 2개의 데이터

input_dim=3: 입력 벡터 X의 차원 3, image.png

```python
# midterm, final, added point
X = np.array([[70,85,11], [71,89,18], [50,80,20], [99,20,10], [50,10,10]])
# score
y = np.array([73, 82 ,72, 57, 34])

model = Sequential()
model.add(Dense(1, input_dim=3, activation='linear'))

sgd = optimizers.SGD(learning_rate=0.0001)
model.compile(optimizer=sgd, loss='mse', metrics=['mse'])
model.fit(X, y, epochs=2000)
```

```python
print(model.predict(X))
```

```python
X_test = np.array([[20, 99, 10], [40, 50, 20]])
print(model.predict(X_test))
```

## **2. 다중 로지스틱 회귀**

---



독립 변수 x가 2개 이상일 때 y 결정

Sepal(꽃받침)길이와 Petal(꽃잎)길이에 따라 Speces (A/B) 결정
image.png

Species: A -> 0, B -> 1

```python
X = np.array([[5.1, 3.5], [4.7, 3.2], [5.2, 1.8], [7, 4.1], [5.1, 2.1]])
y = np.array([0, 0, 1, 0, 1])

model = Sequential()
model.add(Dense(1, input_dim=2, activation='sigmoid'))
model.compile(optimizer='sgd', loss='binary_crossentropy', metrics=['binary_accuracy'])

model.fit(X, y, epochs=2000)
```

```python
print(model.predict(X))
```
