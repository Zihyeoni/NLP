## **2. 비용 함수(Cost function) : 평균 제곱 오차(MSE)**

---



머신 러닝은 ***w***와 ***b***를 찾기 위해 "실제값과 가설로부터 얻은 예측값의 오차를 계산하는 식"을 세우고, 이 식의 값을 최소화하는 최적의 ***w***와 ***b***를 찾아냄

**목적 함수(Objective function) = 비용 함수(Cost function) = 손실 함수(Loss function)**

비용 함수는 단순히 오차를 표현하는 것이 아닌, 예측값의 오차를 줄이는 일에 최적화 된 식이어야 함.

회귀 문제의 경우, 주로 **평균 제곱 오차(Mean Squared Error, MSE)** 사용

오차: 주어진 데이터에서 각 ***x***에서의 실제값 ***y***와 위의 직선에서 예측하고 있는 ***H(x)***값의 차이

오차를 줄여가며 ***w***, ***b*** 값을 찾아내기 위해서는 전체 오차의 크기를 구해야 함
