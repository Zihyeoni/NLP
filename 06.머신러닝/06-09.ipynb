#### 소프트맥스 함수의 입력으로 어떻게 바꿀까?

하나의 샘플 데이터에 4개의 독립 변수 x 가짐 (= 4차원 벡터 입력으로 받음)

그런데 소프트맥스의 함수에서 사용되는 벡터의 차원은 분류하고자 하는 클래스의 개수가 되어야함 (3차원 벡터로 변환 필요)

image.png

=> 입력 벡터 z의 차원수만큼 결과값 나오도록 가중치 곱 진행

위의 경우, 총 12개(4 × 3)의 다른 가중치를 가지며 학습 과정에서 점차 오차를 최소화하는 가중치로 변경됨

#### 오차를 어떻게 구할까?

소프트맥스 함수의 출력은 각각 0 ~ 1 값을 가짐 (= 특정 클래스가 정답일 확률)

이 예측값과 비교할 실제값의 표현 방법 필요 (원-핫 벡터)

만약 현재 샘플 데이터의 실제값이 setosa인 경우

> image.png

오차가 0이 되는 경우: 결과(예측값)가 [0 1 0] 일 때

오차를 계산하기 위해 crossentropy 함수 (비용함수) 사용

## **3. 원-핫 벡터의 무작위성**

---



다수의 클래스 분류하는 문제에서는 클래스 개수만큼 숫자 레이블 필요

정수 인코딩: 직관적인 레이블링 방법 {red, green, blue} -> 0, 1, 2

그러나, 정수 인코딩은 일반적인 다중 클래스 분류 문제에 적절하지 않음

ex) Banana, Tomato, Apple -> 1, 2, 3

> image.png

-> 기계는 Banana가 Apple보다 Tomato에 더 가깝다고 이해

일반적인 분류 문제에서의 각 클래스는 순서 의미 X (각 클래스 간의 오차 균등)

따라서 **원-핫 인코딩**을 통해 레이블 인코딩
